```python
data=pd.read_csv("Data/features_3_sec.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data=data.drop(labels=["filename"],axis=1)
```

```python
data.shape
```

```python
data.info()
```

```python
data.isna().sum().sum()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x=data.label)
plt.xticks(rotation=90);
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
scaler=StandardScaler()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y.to_frame()
```

```python
X=np.array(x,dtype=float)
X=scaler.fit_transform(x)
Y=encoder.fit_transform(y)
print(X.shape)
Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=134)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
##Modelling
```

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import (RandomForestClassifier, AdaBoostClassifier, BaggingClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier)
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import (BernoulliNB,GaussianNB, MultinomialNB)
```

```python
pip install catboost
```

```python
from catboost import CatBoostClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=2000)
rfc.fit(x_train,y_train)
rfc_pred = rfc.predict(x_test)
print("Random forest accuracy: {}".format(metrics.accuracy_score(y_test,rfc_pred)))
print()


abc = AdaBoostClassifier()
abc.fit(x_train,y_train)
abc_pred = abc.predict(x_test)
print("Adaboost accuracy: {}".format(metrics.accuracy_score(y_test,abc_pred)))
print()


bag = BaggingClassifier()
bag.fit(x_train,y_train)
bag_pred = bag.predict(x_test)
print("Bagging accuracy: {}".format(metrics.accuracy_score(y_test,bag_pred)))
print()



hgbc = HistGradientBoostingClassifier()
hgbc.fit(x_train,y_train)
hgbc_pred = hgbc.predict(x_test)
print("Hist gradient boosting accuracy: {}".format(metrics.accuracy_score(y_test,hgbc_pred)))
print()


gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb_pred = gnb.predict(x_test)
print("Gaussian NB accuracy: {}".format(metrics.accuracy_score(y_test,gnb_pred)))
print()


CBC = CatBoostClassifier(silent=True,depth=6,iterations=500)
CBC.fit(x_train, y_train)
CBC_pred = CBC.predict(x_test)
print("CBC accuracy: {}".format(metrics.accuracy_score(y_test,CBC_pred)))
```

```python
##Hist Gradient Boosting has the maximum accuracy
```

```python
##Lets try with Deep neural network
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(512,activation="relu",input_shape=(x_train.shape[1],)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(256,activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64,activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10,activation="softmax"),
])
```

```python
print(model.summary())
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=300, batch_size=128)
```

```python
 _, accuracy = model.evaluate(x_test, y_test, batch_size=128)
```

```python
print("Accuracy:",accuracy)
```

```python
##More accuracy then any other classifier
```
